```python
#handling missing values
df.dropna(subset=['Text'], inplace=True)

#lowercasing the data
df['Text'] = df['Text'].str.lower()

# Create a translation table to remove punctuation
translator = str.maketrans('', '', string.punctuation)

# Remove punctuation from the 'Text' column
df['Text'] = df['Text'].apply(lambda x: x.translate(translator))

#removing numbers
df['Text'] = df['Text'].str.replace('\d+', '', regex=True)
#removing url's, emails, and special characters
df['Text'] = df['Text'].str.replace(r'http\S+|www.\S+', '', regex=True)
df['Text'] = df['Text'].str.replace(r'\S+@\S+', '', regex=True)
df['Text'] = df['Text'].str.replace(r'[^\w\s]', '', regex=True)
df['Text'] = df['Text'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', '', str(x)))

df.head(10)
```

```python
#function for removing special characters
def remove_special_characters(text):
  text = re.sub(r'#', ' ', text)
  return re.sub(r'[^A-Za-z0-9\s]+', '', text)

# Function to remove emojis
def remove_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

df['Text'] = df['Text'].apply(remove_special_characters)
df['Text'] = df['Text'].apply(remove_emojis)
df['Text'] = df['Text'].astype(str)

df.head(10)
```

```python
df.to_csv('cleaned_file.csv', index=False)
```

BERT EMBEDDINGS

```python
import pandas as pd
import torch
from transformers import BertModel, BertTokenizer

model = BertModel.from_pretrained('bert-base-uncased')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

def process_batch(batch):
    input_ids = []
    attention_masks = []

    for tweet in batch['Text']:
        if not isinstance(tweet, str) or len(tweet.strip()) == 0:
            tweet = "[PAD]"

        encoded_dict = tokenizer.encode_plus(
            tweet,
            add_special_tokens=True,
            max_length=30,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )

        input_ids.append(encoded_dict['input_ids'])
        attention_masks.append(encoded_dict['attention_mask'])

    input_ids = torch.cat(input_ids, dim=0).to(device)
    attention_masks = torch.cat(attention_masks, dim=0).to(device)

    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_masks)

    embeddings = outputs[0]
    cls_embeddings = embeddings[:, 0, :].cpu().numpy()

    return cls_embeddings

chunk_size = 500
chunks = pd.read_csv('cleaned_file.csv', encoding='latin1', chunksize=chunk_size)

processed_data = pd.DataFrame()

for chunk in chunks:

    cls_embeddings = process_batch(chunk)
    chunk['embedding'] = list(cls_embeddings)
    processed_data = pd.concat([processed_data, chunk])

processed_data.to_csv('bert_embeddings.csv', index=False)
print("Embedding Shape:", processed_data.shape)
print(processed_data.head(10))
```

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load the LLaMA tokenizer and model with optimized settings
tokenizer = AutoTokenizer.from_pretrained('meta-llama/Llama-2-7b-hf')
model = AutoModelForCausalLM.from_pretrained(
    'meta-llama/Llama-2-7b-hf',
    torch_dtype=torch.float16,
    device_map="auto"
)
model.gradient_checkpointing_enable()

if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': tokenizer.eos_token})

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def process_batch(batch):
    input_ids = []
    attention_masks = []

    for text in batch['Text']:
        if not isinstance(text, str) or len(text.strip()) == 0:
            text = "<pad>"

        encoded_dict = tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=20,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )

        input_ids.append(encoded_dict['input_ids'])
        attention_masks.append(encoded_dict['attention_mask'])

    input_ids = torch.cat(input_ids, dim=0).to(device)
    attention_masks = torch.cat(attention_masks, dim=0).to(device)

    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_masks)

    embeddings = outputs[0]
    sum_embeddings = torch.sum(embeddings, dim=1)
    sum_mask = torch.clamp(attention_masks.sum(dim=1, keepdim=True), min=1e-9)
    mean_pooled_embeddings = (sum_embeddings / sum_mask).cpu().numpy()

    return mean_pooled_embeddings

chunk_size = 20
chunks = pd.read_csv('cleaned_file.csv', encoding='latin1', chunksize=chunk_size)

processed_data = pd.DataFrame()

for chunk in chunks:
    cls_embeddings = process_batch(chunk)
    chunk['embedding'] = list(cls_embeddings)
    processed_data = pd.concat([processed_data, chunk])
    torch.cuda.empty_cache()

processed_data.to_csv('llama_embeddings.csv', index=False)
print(processed_data.head(10))
```

```python
!pip install pandas torch numpy
!pip install -U scikit-fuzzy
!pip install fuzzysets
import fuzzysets as fs
import pandas as pd
import torch
import torch.nn as nn
import numpy as np
import skfuzzy as fuzz

# Load embeddings from CSV files, dynamically detecting numeric columns
def load_embeddings(file_path, embedding_columns=None, exclude_patterns=None):
    df = pd.read_csv(file_path)
    if embedding_columns is None:
        # Detect numeric columns
        numeric_cols = df.select_dtypes(include=[np.number]).columns

        # Exclude columns matching specified patterns
        exclude_patterns = exclude_patterns or []
        exclude_columns = [
            col for col in numeric_cols if any(pattern in col for pattern in exclude_patterns)
        ]
        embedding_columns = [col for col in numeric_cols if col not in exclude_columns]
        print(f"Detected embedding columns (excluding patterns {exclude_patterns}): {embedding_columns}")

    embeddings = df[embedding_columns].apply(pd.to_numeric, errors='coerce').fillna(0)
    embeddings_tensor = torch.tensor(embeddings.values, dtype=torch.float32)

    return df, embeddings_tensor, embedding_columns

exclude_patterns = ["_Hate"]
bert_df, bert_embeddings, embedding_columns = load_embeddings(
    "bert_embeddings.csv", exclude_patterns=exclude_patterns
)
print(len(bert_embeddings))
print("bert_embeddings")
llama_df, llama_embeddings, _ = load_embeddings(
    "llama_embeddings.csv", embedding_columns=embedding_columns
)
print(len(llama_embeddings))
print("llama_embeddings")
# Ensure dimensions match
assert bert_embeddings.shape == llama_embeddings.shape, "Embedding dimensions must match!"

# Define Attention Mechanism
class AttentionCombiner(nn.Module):
    def __init__(self, embedding_dim):
        super(AttentionCombiner, self).__init__()
        self.query = nn.Linear(embedding_dim, embedding_dim)
        self.key = nn.Linear(embedding_dim, embedding_dim)
        self.value = nn.Linear(embedding_dim, embedding_dim)
        self.softmax = nn.Softmax(dim=-1)

    def forward(self, bert_emb, llama_emb):
        # Compute attention scores
        query = self.query(bert_emb)
        key = self.key(llama_emb)
        value = self.value(llama_emb)
        print((bert_emb.shape[0]))

        attention_scores = torch.matmul(query, key.transpose(-2, -1)) / np.sqrt(bert_emb.shape[1])
        attention_weights = self.softmax(attention_scores)
        #print(attention_weights)

        fuzzyfied = np.array(fuzz.membership.sigmf(bert_emb, 0.1, 0.9))

        fuzzyfied_ll = np.array(fuzz.membership.sigmf(llama_emb, 0.1, 0.9))
        fuzzyalgsum = torch.from_numpy(np.maximum(0, fuzzyfied*fuzzyfied_ll))
        #fuzzyalgsum1 = fs.t_norm(fuzzyfied, fuzzyfied_ll)
        #fuzzyalgsum = (fuzz.fuzzymath.fuzzy_add(bert_emb, fuzzyfied, llama_emb, fuzzyfied_ll))
        #print(fuzzyalgsum1)
        #print(len(fuzzyalgsum1))

        # Apply attention weights to LLaMA embeddings and combine
        #combined_embedding = bert_emb + torch.matmul(attention_weights, value)
        combined_embedding = bert_emb + torch.matmul(attention_weights, fuzzyalgsum)
        return combined_embedding

# Initialize attention combiner
embedding_dim = bert_embeddings.shape[1]
attention_combiner = AttentionCombiner(embedding_dim)

combined_embeddings = attention_combiner(bert_embeddings, llama_embeddings)

combined_embeddings_df = pd.DataFrame(combined_embeddings.detach().numpy(), columns=embedding_columns)
output_df = bert_df.copy()
output_df.update(combined_embeddings_df)

output_df.to_csv("combined_embeddings.csv", index=False)
```

ALiBi CLASSIFICATION

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset, random_split
from sklearn.metrics import classification_report, accuracy_score
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

embeddings_df = pd.read_csv('combined_embeddings.csv')
data_df = pd.read_csv('cleaned_file.csv')

embeddings_df['embedding'] = embeddings_df['embedding'].apply(lambda x: np.fromstring(x.strip('[]'), sep=' '))

label_columns = ['United_States_Hate', 'Australia_Hate', 'United_Kingdom_Hate',
                 'South_Africa_Hate', 'Singapore_Hate']
labels = data_df[label_columns].values
embedding_ids = data_df['ID'].values

embedding_dict = {row['ID']: row['embedding'] for _, row in embeddings_df.iterrows()}

class EmbeddingDataset(Dataset):
    def __init__(self, embeddings_dict, labels, embedding_ids):
        self.embeddings_dict = embeddings_dict
        self.labels = labels
        self.embedding_ids = embedding_ids

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        emb_id = self.embedding_ids[idx]
        embedding = self.embeddings_dict[emb_id]
        label = self.labels[idx]
        return torch.tensor(embedding, dtype=torch.float32), torch.tensor(label, dtype=torch.float32), emb_id

dataset = EmbeddingDataset(embedding_dict, labels, embedding_ids)
train_size = int(0.9 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)

embedding_dim = embeddings_df['embedding'].iloc[0].shape[0]
num_classes = len(label_columns)

nhead = max([i for i in range(1, embedding_dim + 1) if embedding_dim % i == 0 and i <= embedding_dim])
print(f"Using embedding_dim: {embedding_dim} and nhead: {nhead}")

class ALiBiTransformerClassifier(nn.Module):
    def __init__(self, embedding_dim, num_classes, nhead):
        super(ALiBiTransformerClassifier, self).__init__()

        self.transformer_layer = nn.TransformerEncoderLayer(
            d_model=embedding_dim,
            nhead=nhead,
            dropout=0.2
        )
        self.transformer = nn.TransformerEncoder(self.transformer_layer, num_layers=4)
        self.fc = nn.Linear(embedding_dim, num_classes)

    def forward(self, x):
        x = x.unsqueeze(1)
        x = self.transformer(x)
        x = x.squeeze(1)
        x = self.fc(x)
        return x

model = ALiBiTransformerClassifier(embedding_dim, num_classes, nhead).to(device)

# Calculate pos_weight and define criterion (to deal with class imbalance)
pos_weight = torch.tensor((labels == 0).sum(axis=0) / (labels == 1).sum(axis=0)).to(device)
criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)

# Optimizer and learning rate scheduler
optimizer = torch.optim.Adam(model.parameters(), lr=0.1)  # Lower learning rate to avoid divergence
scheduler = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr=1e-4,
                                                steps_per_epoch=len(train_loader), epochs=10)

def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=10):
    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        for embeddings, labels, _ in tqdm(train_loader):
            embeddings, labels = embeddings.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(embeddings)
            loss = criterion(outputs, labels)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            scheduler.step()
            train_loss += loss.item()
        train_loss /= len(train_loader)

        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {train_loss:.4f}')

    model.eval()
    val_preds, val_labels = [], []
    with torch.no_grad():
        for embeddings, labels, _ in val_loader:
            embeddings, labels = embeddings.to(device), labels.to(device)
            outputs = model(embeddings)
            preds = (torch.sigmoid(outputs) > 0.3).int().cpu().numpy()
            val_preds.extend(preds)
            val_labels.extend(labels.cpu().numpy())

    print("\nFinal Evaluation on Validation Set:")
    report = classification_report(val_labels, val_preds, target_names=label_columns, output_dict=True, zero_division=1)
    print("Classification Report:")
    print(pd.DataFrame(report).transpose())

    all_preds_df = pd.DataFrame(val_preds, columns=label_columns)
    all_preds_df.to_csv("final_predictions.csv", index=False)

train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=10)
```
